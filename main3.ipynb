```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('best_buy_laptops_2024.csv')
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
# Calculate the mean of the 'offers/price' column
mean_price = df['offers/price'].mean()

# Fill missing values with the mean
df['offers/price'].fillna(mean_price, inplace=True)
```

```python
# Plot 1: Histogram of 'offers/price'
plt.figure(figsize=(10, 6))
plt.hist(df['offers/price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot 2: Scatter plot of 'aggregateRating/ratingValue' vs 'offers/price'
plt.figure(figsize=(10, 6))
plt.scatter(df['aggregateRating/ratingValue'], df['offers/price'], color='orange')
plt.title('Rating vs Price')
plt.xlabel('Rating')
plt.ylabel('Price')
plt.show()
```

```python
# Plot 3: Correlation heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Dataset')
plt.show()
```
